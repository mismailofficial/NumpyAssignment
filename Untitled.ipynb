```python
import numpy as np
import cv2 as c
import os
```

```python
def resizeImage(onlyFiles, lenImage, images):
    for i in range(lenImage):
        img = c.imread(onlyFiles[i], 1)
        images[i] = c.resize(img, (200, 200))
    return images
```

```python
def main():
    root = r'photos/'
    try:
        onlyFiles = next(os.walk("photos"))[2]
        lenImage = len(onlyFiles)
        images = np.zeros([lenImage, 200, 200, 3])
        for i in range(lenImage):
            img = c.imread(root+onlyFiles[i], 1)
            images[i] = c.resize(img, (200, 200))
        return images
    except:
        print("Images not found")
```

```python
print(main())
```
